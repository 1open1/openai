```python
import requests
import json

# üîê Set your OpenRouter API key
API_KEY = "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"

# üß† Text input to analyze
text_input = input("Enter text to analyze: ")

# Step 1Ô∏è‚É£: Format the input as a single prompt (OpenRouter chat format)
prompt = f"""
You are a content filtering assistant.
Classify the following text as one of the following:
0 = Safe
1 = Sensitive
2 = Unsafe

Text: "{text_input}"
Respond only with the number (0, 1, or 2).
"""

# Step 2Ô∏è‚É£: Define endpoint and headers for OpenRouter
url = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json",
}

# Step 3Ô∏è‚É£: Build the request body
body = {
    "model": "openrouter/auto",  # Auto-selects best available model
    "messages": [
        {"role": "system", "content": "You are a strict content moderation model that classifies text safety."},
        {"role": "user", "content": prompt}
    ],
    "temperature": 0.0
}

# Step 4Ô∏è‚É£: Send the request
response = requests.post(url, headers=headers, json=body)
data = response.json()

# Step 5Ô∏è‚É£: Process and display the result
if response.status_code == 200:
    classification = data["choices"][0]["message"]["content"].strip()
    print("\nClassification result:")
    print("0 = Safe | 1 = Sensitive | 2 = Unsafe")
    print(f"Text: {text_input}")
    print(f"‚Üí Label: {classification}")
else:
    print(f"Error: {response.status_code}")
    print(data)
```

```python
import requests
import json

API_KEY = "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"
url = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}


prompt_text = """Dumb Joke: Parallel lines have so much in common.
It‚Äôs a shame they‚Äôll never meet.
###
Dumb Joke: Someone stole my mood ring.
I don‚Äôt know how I feel about that.
###
Dumb Joke:"""

body = {
    "model": "openrouter/auto",
    "messages": [
        {"role": "system", "content": "You are a silly dumb joke generator that always completes the 'Dumb Joke:' line."},
        {"role": "user", "content": prompt_text}
    ],
    "temperature": 0.8,
    "max_tokens": 60
}

response = requests.post(url, headers=headers, json=body)
data = response.json()

if response.status_code == 200:
    print("\nDumb Joke \n")
    print(data["choices"][0]["message"]["content"].strip())
else:
    print("Error:", response.status_code)
    print(data)
```
